```python
filling_list=df['Gender'].dropna().tolist()
df['Gender']=df['Gender'].fillna(pd.Series(np.random.choice(filling_list,size=len(df.index))))
filling_lis=df['Married'].dropna().tolist()
df['Married']=df['Married'].fillna(pd.Series(np.random.choice(filling_lis,size=(len(df.index)))))
filling_lst=df['Dependents'].dropna().tolist()
df['Dependents']=df['Dependents'].fillna(pd.Series(np.random.choice(filling_lst,size=len(df.index))))
fillig_lst=df['Self_Employed'].dropna().tolist()
df['Self_Employed']=df['Self_Employed'].fillna(pd.Series(np.random.choice(fillig_lst,size=len(df.index))))
filling_l=df['LoanAmount'].dropna().tolist()
df['LoanAmount']=df['LoanAmount'].fillna(pd.Series(np.random.choice(filling_l,size=len(df.index))))
filling=df['Loan_Amount_Term'].dropna().tolist()
df['Loan_Amount_Term']=df['Loan_Amount_Term'].fillna(pd.Series(np.random.choice(filling,size=len(df.index))))
filling_=df['Credit_History'].dropna().tolist()
df['Credit_History']=df['Credit_History'].fillna(pd.Series(np.random.choice(filling_,size=len(df.index))))
```
